```python
import spacy
import string
from collections import defaultdict
from heapq import nlargest

# Load the spaCy model and stopwords
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.stop_words.STOP_WORDS
```

```python
text_file = open("/content/spaceX_DP.txt", 'r', encoding='utf-8')
text = text_file.read()
print(text)
```

```python
word_frequencies= defaultdict(int)
words = nlp(text)

for word in words:
    # Check if the word is not a stop word and not punctuation
    if word.text.lower() not in stop_words and word.text.lower() not in string.punctuation:
        word_frequencies[word.text.lower()] += 1  # Converting the text to lowercase , just for consistency


for word, frequency in word_frequencies.items():
    print(f'{word}: {frequency}')
```

```python
# Find the maximum frequency in the word frequencies
maximum_frequency = max(word_frequencies.values())
for word in word_frequencies:
    word_frequencies[word] = word_frequencies[word] / maximum_frequency
for word, freq in word_frequencies.items():
    print(f"{word}: {freq}")
```

```python
# Tokenizing the sentences to score them based on the number of words
sentence_scores = {}
sentences = [sent for sent in words.sents]

for i in sentences:
    for word in i:
        if word.text.lower() in word_frequencies:
            if id in sentence_scores:
                sentence_scores[i] += word_frequencies[word.text.lower()]
            else:
                sentence_scores[i] = word_frequencies[word.text.lower()]

```

```python

from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

sentiment = analyzer.polarity_scores(text)


print("Compound Sentiment Score:", sentiment['compound'])
print("Negative Sentiment Score:", sentiment['neg'])
print("Neutral Sentiment Score:", sentiment['neu'])
print("Positive Sentiment Score:", sentiment['pos'])


if sentiment['compound'] >= 0.05:
    print("The text is generally positive and may be subjective.")
    overall_sentiment = "Positive"
elif sentiment['compound'] <= -0.05:
    print("The text is generally negative and may be subjective.")
    overall_sentiment = "Negative"
else:
    print("The text is relatively neutral and may be more objective.")
    overall_sentiment = "Neutral"


```
