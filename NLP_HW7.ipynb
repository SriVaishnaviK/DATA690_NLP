```python
text_file = open("/content/Burbank.txt", 'r', encoding='utf-8')
text = text_file.read()
print(text)
```

```python
from textblob import TextBlob

blob = TextBlob(text)

# Get the sentiment polarity
sent_polarity = blob.sentiment.polarity

negative_prob= (1 - sent_polarity) / 2  # To convert polarity to a probability

print(f"Probability that the sentiment in thetext is going to negative is {negative_prob:.2f}")
```

**ECERCISE 2**

```python
subjectivity = blob.sentiment.subjectivity
print(f"Overall Sentiment: {sent_polarity:.2f}")
print(f"Subjectivity: {subjectivity:.2f}")
```

**EXERCISE 3**

```python
#Extracting key topics as nouns
from textblob import TextBlob, Word

blob = TextBlob(text)

nouns = []
for word, tag in blob.tags:
    word = Word(word)
    if tag == 'NN' or tag == 'NNS':
        nouns.append(word)
print("Nouns:", nouns)
```

**EXERCISE 4**

```python
import pandas as pd

amazon_data = pd.read_csv('amazon_cells_labelled.txt', sep='\t', header=None, names=['Review', 'Label'])
imdb_data = pd.read_csv('imdb_labelled.txt', sep='\t', header=None, names=['Review', 'Label'])
yelp_data = pd.read_csv('yelp_labelled.txt', sep='\t', header=None, names=['Review', 'Label'])


amazon_data['Company'] = 'Amazon'
imdb_data['Company'] = 'IMDb'
yelp_data['Company'] = 'Yelp'

# Concatenating the dataframes
comb_data = pd.concat([amazon_data, imdb_data, yelp_data], ignore_index=True)

```

```python

print("Structure of combined data:")
print(comb_data.head())

comb_data.to_csv('Sentiment_Analysis_Dataset.csv', index=False)

print("\nColumns in comb_data:",comb_data.columns)

print("\nNull values in comb_data:",comb_data.isnull().sum())

```

```python
import pandas as pd
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import string

nlp = spacy.load('en_core_web_sm')

# Stopwords and punctuation lists
stopwords = list(STOP_WORDS)
punctuations = string.punctuation

for sentence in comb_data['Review']:
    doc = nlp(sentence)#Parsing
    tokens = []
    for token in doc:
        if (token.text not in stopwords) and (token.text not in punctuations):
            tokens.append(token.text)
    print(tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.base import TransformerMixin
import re
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(comb_data['Review'], comb_data['Label'], test_size=0.2)


class predictors(TransformerMixin):
  def transform(self, X, **transform_params):
    return [clean_text(text) for text in X]

  def fit(self, X, y=None, **fit_params):
    return self

  def get_params(self, deep=True):
    return {}


def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = ' '.join(text.split())
    return text

#Pipeline
pipe_countvect = Pipeline([
    ('clean', predictors()),
    ('tfidf', TfidfVectorizer()),
    ('svc', LinearSVC())
])

pipe_countvect.fit(X_train, y_train)

pred = pipe_countvect.predict(X_test)

for x, y in zip(X_test, pred):
  print(x, "Prediction→", y)

```

```python
print("Test accuracy:", accuracy_score(y_test, pred))
print("Train accuracy:", accuracy_score(y_train, pipe_countvect.predict(X_train)))
```
